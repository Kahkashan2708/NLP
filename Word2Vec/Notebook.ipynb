```python
import gensim
from gensim.models import Word2Vec, KeyedVectors
```

```python
path = "/kaggle/input/googlenewsvectorsnegative300/GoogleNews-vectors-negative300.bin"
model = KeyedVectors.load_word2vec_format(path, binary=True, limit=500000)
```
